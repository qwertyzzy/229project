```python
import os
import pandas as pd
import numpy as np
import sqlite3
```

```python
df = pd.read_csv("data.csv")
print(df)
```

```python
youtube_id = list(df["youtube_id"])
timestamp_ms = list(df["timestamp_ms"])
class_id = list(df["class_id"])
class_name = list(df["class_name"])
object_id = list(df["object_id"])
object_presence = list(df["object_presence"])
xmin = list(df["xmin"])
ymin = list(df["ymin"])
xmax = list(df["xmax"])
ymax = list(df["ymax"])
```

```python
actual_class = []
for i in range(len(youtube_id)):
    if object_presence[i] == 'present':
        actual_class.append(class_id[i])
    else:
        actual_class.append(-1)
print(actual_class)
```

```python
db_path = 'cluster.db'
con = sqlite3.connect(db_path)
cur = con.cursor()

for i in range(len(youtube_id)):
    cur.execute('''
                INSERT INTO main (video_id, frame, image_id, class, object_id, object_presence, xmin, 
                xmax, ymin, ymax, actual_class, train)
                VALUES
                (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
                ''', (youtube_id[i], timestamp_ms[i], i, class_id[i], object_id[i], object_presence[i], xmin[i], xmax[i], ymin[i], ymax[i], actual_class[i], 0))
con.commit()
```
