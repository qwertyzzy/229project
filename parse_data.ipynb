```python
def file_exists(df):
    path = root + str(df[2]) + '/' + str(df[0]) + '_' + str(df[1]) + '_' + str(df[2]) + '_' + str(df[4]) + '.jpg'
    # print(path)
    return os.path.isfile(path)
```

```python
drop_index = []
for index, row in df.iterrows():
    if not file_exists(row):
        drop_index.append(index)
print(drop_index)
```

```python
partial_df = df.drop(df.index[drop_index])
```

```python
print(df.shape)
print(partial_df.shape)
partial_df = partial_df[:200000]
print(partial_df.shape)
```
